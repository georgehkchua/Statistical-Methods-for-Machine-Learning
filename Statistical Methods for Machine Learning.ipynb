```python
# Statistical Methods for Machine Learning
# Chapter 4 - Gaussian and Summary Stats
from numpy import arange, mean, var, std
from matplotlib import pyplot
from scipy.stats import norm
from numpy.random import seed, randn
x_axis = arange(-3, 3, 0.001)
seed(1)
data = 5 * randn(10000) + 50
result_mean = mean(data)
result_var = var(data)
result_std = std(data)
pyplot.plot(x_axis, norm.pdf(x_axis, 0, 0.25))
pyplot.plot(x_axis, norm.pdf(x_axis, 0, 0.5))
pyplot.plot(x_axis, norm.pdf(x_axis, 0, 1))
pyplot.plot(x_axis, norm.pdf(x_axis, 0, 2))
pyplot.show()
pyplot.hist(data, bins=500)
pyplot.show()
print('Mean: %.3f' %result_mean)
print('Variance: %.3f' %result_var)
print('Standard Deviation: %.3f' %result_std)
```
